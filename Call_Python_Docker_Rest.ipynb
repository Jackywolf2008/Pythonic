```python
import requests as req
import json
from pprint import pprint
```

```python
containerURL = 'http://summer.edmt.sashq-d.openstack.sas.com:8889'
publishEndpoint = containerURL + '/models'
scoreEndpoint = containerURL + '/executions'
queryEndpoint = containerURL + '/query/'
```

```python
# publish the python model

modelZipFile = 'dtree_cas.zip'


with open(modelZipFile,'rb') as f:
    mdl={'file':(modelZipFile,f,'multipart/form-data')}
    response = req.post(publishEndpoint, files=mdl)

pprint(response)

```

```python
jobDefinitionId = json.loads(response.content.decode('utf-8'))['id']
print(jobDefinitionId)
```

```python
# run the published model
inputCsvFile = 'hmeq.csv'
with open(inputCsvFile,'rb') as inputf:
    inData={'file':(inputCsvFile, inputf, 'multipart/form-data')}
    response = req.post(scoreEndpoint, files=inData, data={'job_definition_id':jobDefinitionId})
pprint(response)    
```

```python
testId = json.loads(response.content.decode('utf-8'))['id']
print(testId)
```

```python

response = req.get(queryEndpoint + "/".join([jobDefinitionId, testId]))
pprint(response)
```
